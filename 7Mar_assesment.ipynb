The three measures of central tendency are:

Mean: The mean is the most common measure of central tendency. It is calculated by adding up all the values in a dataset and then dividing by the total number of values. The mean is sensitive to extreme values, also known as outliers.

Median: The median is the middle value in a dataset when the values are arranged in order from lowest to highest (or highest to lowest). If there are an even number of values in the dataset, the median is the average of the two middle values. The median is less sensitive to outliers than the mean.

Mode: The mode is the value that appears most frequently in a dataset. A dataset can have one or more modes, or no mode at all. The mode is often used for categorical data or when the data is not numerical.

The mean, median, and mode are all measures of central tendency, which means they provide a way to describe the center of a dataset. However, they differ in how they are calculated and what information they provide.

Measures of dispersion, such as range, variance, and standard deviation, are used to describe the spread or variability of a dataset.

The range is the difference between the largest and smallest values in a dataset, and it provides a simple measure of dispersion. However, it can be sensitive to outliers, which are extreme values that can skew the range.

The variance and standard deviation are more robust measures of dispersion that take into account the deviation of each data point from the mean. The variance is the average squared deviation of each data point from the mean, while the standard deviation is the square root of the variance. The standard deviation is often preferred over the variance because it has the same units as the data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create the dataset
data = np.array([68, 75, 80, 82, 85, 87, 90, 92, 95, 98])

# Calculate the range, variance, and standard deviation
range = np.max(data) - np.min(data)
variance = np.var(data, ddof=1)
std_dev = np.std(data, ddof=1)

# Create the histogram
plt.hist(data, bins=5, edgecolor='black')

# Add the range, variance, and standard deviation to the plot
plt.axvline(x=np.mean(data)-range/2, color='r', linestyle='--', label='Range')
plt.axvline(x=np.mean(data)-std_dev, color='g', linestyle='--', label='Standard Deviation')
plt.axvline(x=np.mean(data), color='b', linestyle='-', label='Mean')
plt.axvline(x=np.mean(data)+std_dev, color='g', linestyle='--')
plt.axvline(x=np.mean(data)+range/2, color='r', linestyle='--')

# Add labels and a legend
plt.xlabel('Test Score')
plt.ylabel('Frequency')
plt.title('Distribution of Test Scores')
plt.legend()

# Show the plot
plt.show()
```

Covariance and correlation are both measures of the relationship between two variables, but they differ in a few key ways:

1. Definition: Covariance measures the degree to which two variables vary together. In other words, it measures the extent to which a change in one variable is associated with a change in the other variable. Correlation, on the other hand, measures the strength and direction of the linear relationship between two variables.

2. Scale: Covariance can take on any value, positive or negative, and its magnitude is not standardized, meaning it is dependent on the scale of the variables being measured. Correlation, however, always takes on a value between -1 and 1, and its magnitude represents the strength of the relationship between the two variables.

3. Interpretation: The interpretation of covariance and correlation also differs. A positive covariance indicates that two variables tend to increase or decrease together, while a negative covariance indicates that they tend to vary in opposite directions. A correlation of +1 or -1 indicates a perfect linear relationship between two variables, while a correlation of 0 indicates no linear relationship between them.

4. Units: Covariance is in units that are the product of the units of the two variables being measured, whereas correlation is unitless since it is a standardized measure.

Outliers are data points that lie far away from the rest of the data in a dataset. They can have a significant impact on measures of central tendency and dispersion. Here's how:

1. Measures of central tendency: Outliers can significantly affect the mean (average) of a dataset because they pull the mean towards them. For example, if we have a dataset of 9 values: 1, 2, 3, 4, 5, 6, 7, 8, and 200, the mean would be 24.6. However, the outlier value of 200 skews the mean, making it higher than what it would have been if the outlier was not present. In contrast, outliers have little effect on the median, which is the middle value in a dataset. Therefore, if a dataset has outliers, it may be more appropriate to use the median as the measure of central tendency.

2. Measures of dispersion: Outliers can also significantly affect measures of dispersion such as the range and standard deviation. The range is the difference between the maximum and minimum values in a dataset. If an outlier is present, it can increase the range substantially. Similarly, the standard deviation measures the spread of data around the mean, and an outlier can increase the standard deviation, making it an unreliable measure of dispersion. Therefore, when outliers are present, it may be more appropriate to use alternative measures of dispersion such as the interquartile range (IQR) or median absolute deviation (MAD), which are less sensitive to outliers.

For example, let's say we have the following dataset of 10 exam scores: 60, 70, 75, 80, 85, 90, 95, 100, 105, and 200. The outlier value of 200 has a significant effect on the measures of central tendency and dispersion:

i) Mean: The mean score is 92.5, which is higher than what it would have been if the outlier was not present.
ii )Median: The median score is 87.5, which is less affected by the outlier.
iii) Range: The range is 140, which is substantially higher than what it would have been without the outlier.
iv) Standard deviation: The standard deviation is 43.36, which is significantly higher than what it would have been without the outlier.
